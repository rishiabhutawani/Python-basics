10. What is difference between implicit and explicit type casting ?
->  Implicit Type Casting
Done automatically by Python.
Occurs when no data loss will happen during conversion.
Happens during expressions involving mixed data types.

->. Explicit Type Casting
Done manually by the programmer.
You specify the desired data type using functions like int(), float(), str(), etc.
Used when you want to control or force the conversion.

```python
# write a function to demonstrate the behaviour of mutable and immutable arguments.
def modify_arguments(immutable_arg, mutable_arg):
    immutable_arg += 10
    print("Inside function - immutable_arg:", immutable_arg)
    mutable_arg.append(100)
    print("Inside function - mutable_arg:", mutable_arg)
num = 5
numbers = [1, 2, 3]
print("Before function call:")
print("num:", num)
print("numbers:", numbers)
modify_arguments(num, numbers)
print("\nAfter function call:")
print("num:", num)
print("numbers:", numbers)

```

```python
# Write a python program to convert user input from string to integer, float, and boolean types.
user_input = input("Enter a value: ")
try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to integer.")
try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to float.")
bool_value = user_input.strip().lower() not in ["", "false", "0", "no"]
print("Boolean value:", bool_value)
```

```python
# Write a python program to find the sum of all even numbers between 1 and 50.
even_sum = 0

for i in range(1, 51):
    if i % 2 == 0:
        even_sum += i

print("Sum of even numbers from 1 to 50 is:", even_sum)

```

```python
# Write a program to reverse a string using a while loop.
input_string = input("Enter a string: ")
reversed_string = ""
index = len(input_string) - 1
while index >= 0:
    reversed_string += input_string[index]
    index -= 1
```

```python
# Write a python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
factorial = 1
current = 1
while current <= num:
    factorial *= current
    current += 1
print(f"The factorial of {num} is {factorial}")
```
